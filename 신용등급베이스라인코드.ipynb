```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 각 직업에 대해 신용 점수의 빈도수를 계산
occupation_score_counts = df.groupby(['Occupation', 'Credit_Score']).size().reset_index(name='Count')

# 각 직업별 총 빈도수 계산
total_counts = occupation_score_counts.groupby('Occupation')['Count'].transform('sum')

# 비율 계산
occupation_score_counts['Probability'] = occupation_score_counts['Count'] / total_counts

# 시각화 준비
fig, ax = plt.subplots(figsize=(12, 8))

# 고유 직업과 신용 점수 가져오기
occupations = occupation_score_counts['Occupation'].unique()
credit_scores = sorted(occupation_score_counts['Credit_Score'].unique())  # 신용 점수를 정렬

# 막대 색상 설정
colors = sns.color_palette('viridis', len(credit_scores))

# 누적 막대 그래프 생성
bottoms = pd.Series(0, index=occupation_score_counts['Occupation'].unique())
for score in credit_scores:
    subset = occupation_score_counts[occupation_score_counts['Credit_Score'] == score]
    ax.bar(subset['Occupation'], subset['Probability'] * 100,  # 비율을 백분율로 변환
           label=str(score), color=colors[credit_scores.index(score)],  # 색상 매핑 수정
           bottom=bottoms.loc[subset['Occupation']])
    bottoms.loc[subset['Occupation']] += subset['Probability'] * 100  # 업데이트된 바텀값

# Y축 최대값 조정
ax.set_ylim(0, 100)  # 백분율 값에 맞게 100%로 설정

# 텍스트 추가
for occupation in occupations:
    subset = occupation_score_counts[occupation_score_counts['Occupation'] == occupation]
    cumulative_bottom = 0
    for _, row in subset.iterrows():
        value = row['Probability'] * 100
        ax.text(occupation, cumulative_bottom + value / 2,
                f'{value:.1f}%', ha='center', va='center', fontsize=8)
        cumulative_bottom += value

# 레이아웃 설정
ax.set_title('Credit Score Distribution by Occupation (100% Stacked)')
ax.set_xlabel('Occupation')
ax.set_ylabel('Percentage (%)')
ax.legend(title='Credit Score')
ax.tick_params(axis='x', rotation=90)

# X축 레이블 설정
ax.set_xticks(list(occupation_score_counts['Occupation'].unique()))
ax.set_xticklabels(list(occupation_score_counts['Occupation'].unique()), rotation=90)

plt.tight_layout()
plt.show()
```

#### 클러스터링
나이 , 신용카드 이자율 빈도 , 신용카드 조회 횟수

완만해지는 지점 4를 선택하여 최적의 k를 4로 지정하겠다.

- Cluster 0과 3은 Good 등급이 많이 없다 0과 1은 poor이 비중이 많은 편이며
이에 따른 등급 예측과 이에 따른 대출 패키지를 제공해야한다

- Cluster 1,2는 Good 등급이 많은 편이고 Standarnd 또한 많고 poor이 적다
그렇기에 standard 등급을 Good 등급으로 올릴 수 있는 패키지 대출 상품을 제공 필요


- 빈도 높은 순

- bad : Cluster 0의 연령대는  30 ~ 40대  이자율은 17 ~ 20  조회 7 ~ 11회
- bad : Cluster 3의 연령대는  20대	    이자율 20 후반 조회 6~12회


- good :Cluster 1의 연령대는  20 ~ 30대  이자율 5 ~ 12	조회 4회가 가장 많음 0 ~ 6회
- good : Cluster 2의 연령대는  40 ~ 50대  이자율 5 ~ 12	조회 4회가 가장 많음 0 ~ 4회

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt



# 나이 그룹 생성 (구간 나누기)
bins = [0, 9, 19, 29, 39, 49, 59, 100]
labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-100']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# 클러스터링을 위한 피처 선택
X = df[['Interest_Rate', 'Age', 'Num_Credit_Inquiries']]

# 피처 표준화 (Standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 엘보우 기법을 위한 클러스터 수 범위 설정
wcss = []
k_values = range(1, 11)

# KMeans 모델을 각 k값에 대해 적용하고 WCSS 계산
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)  # WCSS 값 저장

# 엘보우 그래프 그리기
plt.plot(k_values, wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.show()
```
